```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os

directory_path = "/global/cfs/cdirs/atlas/jmw464/nishank_pi2/tdd_samples_mod"

def plot_histogram(data, title, xlabel, ylabel, bins=50):
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=bins, edgecolor='black')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

def find_hdf5_files(directory):
    hdf5_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.h5'):
                hdf5_files.append(os.path.join(root, file))
    return hdf5_files

all_quark_charge_bottom = []
all_quark_charge_charm = []

# Find all HDF5 files in the directory and subdirectories
file_paths = find_hdf5_files(directory_path)

if not file_paths:
    print("No HDF5 files found in the directory.")
else:
    for file_path in file_paths:
        if os.path.isfile(file_path):  # Ensure it is a file
            print(f"Processing file: {file_path}")  # Debug print
            try:
                with h5py.File(file_path, 'r') as f:
                    # Check what keys are available
                    print(f"Keys in the file: {list(f.keys())}")  # Debug print
                    
                    # Check datasets in 'jets' and 'tracks'
                    jets_dataset = f['jets']
                    tracks_dataset = f['tracks']
                    
                    # Print attributes of the datasets
                    print(f"Attributes in 'jets' dataset: {list(jets_dataset.attrs.keys())}")
                    print(f"Attributes in 'tracks' dataset: {list(tracks_dataset.attrs.keys())}")
                    
                    # Print shape of the datasets
                    print(f"Shape of 'jets' dataset: {jets_dataset.shape}")
                    print(f"Shape of 'tracks' dataset: {tracks_dataset.shape}")
                    
                    # Assuming the necessary data is within these attributes, adapt accordingly
                    # For now, we just print and understand the structure
                    # This part will change once we know the correct structure
                    
                    # Example of accessing an attribute if it exists
                    if 'quarkCharge' in jets_dataset.attrs:
                        quark_charge = jets_dataset.attrs['quarkCharge']
                        hadron_label = jets_dataset.attrs['HadronConeExclTruthLabelPdgId']
                        
                        # Separate bottom and charm hadrons based on their PDG IDs
                        bottom_hadrons = (abs(hadron_label) // 1000) == 5
                        charm_hadrons = (abs(hadron_label) // 1000) == 4

                        # Filter quark charge for bottom and charm hadrons
                        quark_charge_bottom = quark_charge[bottom_hadrons]
                        quark_charge_charm = quark_charge[charm_hadrons]

                        all_quark_charge_bottom.extend(quark_charge_bottom)
                        all_quark_charge_charm.extend(quark_charge_charm)
                    
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")

# Convert to numpy arrays for plotting
all_quark_charge_bottom = np.array(all_quark_charge_bottom)
all_quark_charge_charm = np.array(all_quark_charge_charm)

# Debug prints to check data
print(f"Number of bottom quark charges: {len(all_quark_charge_bottom)}")
print(f"Number of charm quark charges: {len(all_quark_charge_charm)}")

# Plot histograms
if len(all_quark_charge_bottom) > 0:
    plot_histogram(all_quark_charge_bottom, 'Quark Charge Distribution for Bottom Hadrons', 'Quark Charge', 'Frequency')
else:
    print("No bottom quark charges to plot.")

if len(all_quark_charge_charm) > 0:
    plot_histogram(all_quark_charge_charm, 'Quark Charge Distribution for Charm Hadrons', 'Quark Charge', 'Frequency')
else:
    print("No charm quark charges to plot.")
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import os

def calculate_cut_value(discriminant, target_efficiency=0.7):
    """
    Calculate the cut value for a given target efficiency.
    
    Args:
    discriminant (np.array): Array of discriminant values.
    target_efficiency (float): Target efficiency (default 0.7 for 70%).
    
    Returns:
    float: Cut value that achieves the target efficiency.
    """
    sorted_disc = np.sort(discriminant)[::-1]  # Sort in descending order
    cut_index = int(len(sorted_disc) * target_efficiency)
    return sorted_disc[cut_index]

def apply_working_point(jets_data, discriminant_name, target_efficiency=0.7):
    """
    Apply the working point to the jets data.
    
    Args:
    jets_data (np.array): Structured array of jet data.
    discriminant_name (str): Name of the discriminant column.
    target_efficiency (float): Target efficiency (default 0.7 for 70%).
    
    Returns:
    np.array: Filtered jets data.
    """
    discriminant = jets_data[discriminant_name]
    cut_value = calculate_cut_value(discriminant, target_efficiency)
    return jets_data[discriminant >= cut_value]

def process_file(file_path, discriminant_name, target_efficiency=0.7):
    with h5py.File(file_path, 'r') as f:
        jets_data = f['jets'][:]
        filtered_jets = apply_working_point(jets_data, discriminant_name, target_efficiency)
        hadron_pdg_ids = filtered_jets['HadronConeExclTruthLabelPdgId']
        child_pdg_ids = filtered_jets['HadronConeExclTruthLabelChildPdgId']
    return hadron_pdg_ids, child_pdg_ids

# Directory path containing HDF5 files
dir_path = '/global/cfs/cdirs/atlas/jmw464/nishank_pi2/tdd_samples/'

# Initialize empty lists to store the data
hadron_pdg_ids_all = []
child_pdg_ids_all = []

# Counter to keep track of the number of processed files
file_count = 0

# Specify the discriminant column name
discriminant_name = 'your_discriminant_column_name'  # Replace with your actual discriminant column name

# Iterate over files in the directory
for root, dirs, files in os.walk(dir_path):
    for filename in files:
        if filename.endswith('.h5'):
            file_path = os.path.join(root, filename)
            print(f"Processing file: {file_path}")
            try:
                hadron_pdg_ids, child_pdg_ids = process_file(file_path, discriminant_name, target_efficiency=0.7)
                hadron_pdg_ids_all.extend(hadron_pdg_ids)
                child_pdg_ids_all.extend(child_pdg_ids)
                file_count += 1
            except Exception as e:
                print(f"Failed to process file {file_path}: {e}")
            # Process only the first 15 files
            if file_count >= 15:
                break
    if file_count >= 15:
        break

# Convert to numpy arrays for further processing
hadron_pdg_ids_all = np.array(hadron_pdg_ids_all)
child_pdg_ids_all = np.array(child_pdg_ids_all)

# ... rest of your code for plotting and analysis ...
```
